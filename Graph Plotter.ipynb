<h3>The following snippet contains the code to display the temperature vs time plot for the analog temperature sensor. The text file data_rohan.txt was obtained from the serial monitor while running the code Sensor.ino on the Arduino R3 microcontroller board.</h3>

<h3>A similar plot was done by Aabir for his solar cooker and the resultant plot can be seen in the report</h3>

```python
#Rohan Rajesh Kalbag

import matplotlib.pyplot as plt
import numpy as np

#Changing plot size
plt.rcParams['figure.figsize'] = [18, 10]
plt.rcParams['axes.labelsize'] = 16
plt.rcParams['axes.titlesize'] = 20

file = open("data_rohan.txt","r")
a = file.readlines()
data = []

for i in a:
    x = i.split()
    data += [float(x[2])]

file.close()
time = [ 10*i for i in range(1,len(data)+1)]
timenp = np.array(time)
timenp = timenp/60
datanp = np.array(data)
  
# it reached steady state at 15 min ~ 900s calculating average temp
#from 15 min i.e the 90th reading till the end of the measurement 

avg = 0
initread = 90
while(initread != len(data)-1):
      avg += data[initread]
      initread += 1
avg /=  270 # 270  readings

plt.plot(timenp,datanp,'r')
plt.xlabel("Time in min")
plt.ylabel("Temperature in Â°C of water")
plt.show()
```
